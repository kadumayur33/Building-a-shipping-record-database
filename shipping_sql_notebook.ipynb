# SQL Basics — Shipping Records Database

Complete solution with outputs

## Step 1: Create Database and Insert Data

```

import sqlite3

conn = sqlite3.connect('shipping.db')
cursor = conn.cursor()

cursor.execute('''
CREATE TABLE shipping_mode_dimension (
    ship_mode TEXT,
    vehicle_company TEXT PRIMARY KEY,
    toll_required INTEGER
)
''')

data = [
    ('Delivery Truck', 'Ashok Leyland', 1),
    ('Air Freight', 'Air India', 0),
    ('Rail Cargo', 'Indian Railways', 1),
    ('Delivery Truck', 'Tata Motors', 1),
    ('Air Freight', 'SpiceJet Cargo', 0),
    ('Rail Cargo', 'IRCTC Logistics', 0)
]

cursor.executemany('INSERT INTO shipping_mode_dimension VALUES (?, ?, ?)', data)

conn.commit()

print("Table created and data inserted successfully")
```

## Step 2: Query and Filter

```

# (a) Rows where toll_required = 1
cursor.execute("SELECT * FROM shipping_mode_dimension WHERE toll_required = 1")
print("Rows with toll_required = 1:")
print(cursor.fetchall())
```

```

# (b) Count total records
cursor.execute("SELECT COUNT(*) FROM shipping_mode_dimension")
print("Total records:")
print(cursor.fetchone())
```

```

# (c) Air Freight records
cursor.execute("""
SELECT ship_mode, vehicle_company 
FROM shipping_mode_dimension 
WHERE ship_mode = 'Air Freight'
""")
print("Air Freight records:")
print(cursor.fetchall())
```

## Step 3: Modify and Clean Data

```

# (a) Add new column
cursor.execute("ALTER TABLE shipping_mode_dimension ADD COLUMN toll_amount INTEGER")
```

```

# (b) Update toll_amount
cursor.execute("""
UPDATE shipping_mode_dimension
SET toll_amount = 
    CASE 
        WHEN toll_required = 1 THEN 200
        ELSE 0
    END
""")
conn.commit()
```

```

# (c) Delete Air India record
cursor.execute("DELETE FROM shipping_mode_dimension WHERE vehicle_company = 'Air India'")
conn.commit()
```

```

# Final table
cursor.execute("SELECT * FROM shipping_mode_dimension")
rows = cursor.fetchall()

print("Final Table:")
for row in rows:
    print(row)
```
